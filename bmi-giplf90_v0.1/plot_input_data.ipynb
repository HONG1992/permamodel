This script visualizes input data from the example folder. Note, that measurements correspond to the beginning of the hydrologic year (July 1). 

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_data(input_file,ylabel):
    input_data = np.loadtxt(input_file, skiprows=3,  unpack=False)
    data=input_data[:,1]
    plt.plot(data,'b',linewidth=2.0)
    plt.xlim([0, len(data)]);
    plt.ylabel(ylabel, fontsize=16); 
    plt.tick_params(labelsize=16)
```

```python
fig = plt.figure()
fig.set_figheight(8); fig.set_figwidth(11)

fig.add_subplot(3,1,1)
bound_file='examples/bound.txt'
plot_data(bound_file,'Air temp [C]')
plt.plot(np.zeros(365),'k--',linewidth=2.0)

fig.add_subplot(3,1,2)
snow_file='examples/snow.txt'
plot_data(snow_file,'snow depth [m]')

fig.add_subplot(3,1,3)
rsnow_file='examples/rsnow.txt'
plot_data(rsnow_file,'snow density [kg m-3]')
plt.xlabel('time [days]',  fontsize=16);
fig.tight_layout()
```
